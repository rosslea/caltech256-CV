```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('data/performance.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
loss_train, loss_valid = data['loss_train'][:50],data['loss_valid'][:50]
acc_train, acc_valid = data['acc_train'][:50], data['acc_valid'][:50]
X = np.arange(len(loss_train))
Y1 = np.array(loss_train)
Y2 = np.array(loss_valid)
Y3 = np.array(acc_train)
Y4 = np.array(acc_valid)
```

```python
fig, ax = plt.subplots()
ax.plot(X,Y1,label='loss train')
ax.plot(X,Y2, label='loss valid')
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(X,Y3, label='acc train')
ax.plot(X,Y4, label='acc valid')
plt.legend()
```

```python
with open('data/test_one_performance.pkl', 'rb') as f:
    data = pickle.load(f)
output = data['output']
label = data['label']
```

```python
from pathlib import Path
cls = [i.name for i in list(Path('./256_ObjectCategories').iterdir())]
cls = [(i[:3],i[4:]) for i in cls]
cls.sort(key=lambda x:x[0])
names = [i[1] for i in cls]
```

```python
Y = np.array(output[0])
zpd = list(zip(names, Y))
zpd.sort(key=lambda x:x[1])
X = [i[0] for i in zpd][-11:]
Y = [i[1] for i in zpd][-11:]
```

```python
fig, ax = plt.subplots()
ax.scatter(X,Y)
ax.plot(X,Y)
plt.xticks(rotation=50)
```
